## import stuff...

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh

# Support vector classifier
from sklearn.svm import SVC 

# also define the default font we'll use for figures. 
fig_font = {'fontname':'Arial', 'size':'16'}
```

```python
# load the data...
eeg = np.load('class_demo_eeg.npz')

# get the different arrays like this...
eeg['data']
eeg['sr']
eeg['tx']
eeg['cond_labels']

# shape of data
print('Shape of the big eeg data set: ', eeg['data'].shape)

# and if you want to save some typing, especially because we only have a few variables, you reassign the different arrays like this
data = eeg['data']
sr = eeg['sr']
tx = eeg['tx']
condLabels = eeg['cond_labels']
```

```python
# get the mean over the first dim
mean_data = np.mean(data, axis=0)
print(mean_data)
print("shape:", mean_data.shape)

plt.figure(figsize=(20,10))
plt.plot(tx, mean_data.T) #transposed
plt.xlabel('Time (msec)', **fig_font)
plt.ylabel('Response (mV)', **fig_font)
plt.show()
```

## That gives you an idea about what the pattern of data looks like across electrodes. 
* Next, take the mean across time, so you have a 960 x 20 matrix 
* Use the cond_labels vector to sort the data based on condition (so you have two 480 x 20 matrices)
* Do classification to see if you can decode the experimental condition
* Remember - break up your data into independent training and testing sets
* Set C to a small number if this is taking a long time...

```python
meanDataAcrossTime = np.mean(data, axis=2)
cond0Index = condLabels == 0
cond1Index = condLabels == 1
```

```python
cond0 = meanDataAcrossTime[cond0Index,:] 
cond1 = meanDataAcrossTime[cond1Index,:]

def classifyTwoConditionsEEGDemo(cond0,cond1):
    
    r,c = cond0.shape #finding the rows, columns of the data

    # get the length of the training data (90% of the datapoints per cond)
    trainLength = int(np.round(0.9*r)) 

    #stack the training data from cond0 and cond1 together
    trainStacked = np.vstack((cond0[0:trainLength,:], cond1[0:trainLength,:]))

    #stack the test data from cond0 and cond1 together (last 10% of the datapoints per cond)
    testStacked = np.vstack((cond0[trainLength:,:],cond1[trainLength:,:]))

    #make an index matrix for keeping track of the conds, for...
    trainLabel = np.hstack((np.zeros(trainLength),np.ones(trainLength))) #training
    testLabels = np.hstack((np.zeros(r-trainLength),np.ones(r-trainLength))) #test

    #find the mean for each cond in the training set
    #mean for trials, but not across electrodes
    mean0 = np.mean(trainStacked[trainLabel==0,:], axis=(0)) 
    mean1 = np.mean(trainStacked[trainLabel==1,:], axis=(0))

    # initialize array for classification labels
    classLabels = np.zeros(testStacked.shape[0])
    for i in np.arange(testStacked.shape[0]):
        # compute Euc distance from each point to the mean of each condition
        distTo0 = np.sqrt( np.sum( np.power( (testStacked[i,:]-mean0) , 2) ) )
        distTo1 = np.sqrt( np.sum( np.power( (testStacked[i,:]-mean1) , 2) ) )
        # then figure out which distance is smaller
        # (which cloud our test point is closer to)
        classLabels[i] = np.argmin([distTo0, distTo1])

    # figure out classification accuracy
    # compare the predicted labels to the ground truth
    classAccuracy = np.sum(classLabels==testLabels)/testStacked.shape[0]
    print('classification accuracy:',classAccuracy)
    return classAccuracy

classifyTwoConditionsEEGDemo(cond0,cond1)
```
