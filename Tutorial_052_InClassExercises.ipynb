## What about outliers? Lets define outliers here as > 2 * std away from the mean for each subject
* After you've found the outliers for each subject, replace those values with a np.nan (NaN)

```python
describedDataFrame
```

```python
dcDataFrame
```

```python
def replaceOutliersWithNans(df, howManyStDevs):
    newDf = df.copy()
    descDf = df.describe()
    for i in range(0,len(df)):
        for j in range(0,len(df.T)):
            if abs(df.iloc[i][j] - descDf.loc['mean'][j]) > (howManyStDevs*descDf.loc['std'][j]):
                newDf.iloc[i][j] = np.NaN
    return newDf
```

```python
dcOutliersAreNans = replaceOutliersWithNans(dcDataFrame, 2)
```

```python
dcOutliersAreNans
```
